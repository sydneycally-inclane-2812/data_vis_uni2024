```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
# import polars as pl
first_import = True
```

```python
# data from abs for median wage
country = ['Vietnam', 'China', 'India', 'Philippines', 'US', 'UK', 'Indonesia', 'Thailand', 'Malaysia', 'Singapore', 'Korea']
median_2016 = [456, 374, 785, 746, 832, 707, 573, 449, 721, 678, 491]
median_2021 = [559, 601, 1015, 898, 1020, 813, 757, 657, 900, 936, 754]

# DataFrame
wage = pd.DataFrame({'Country': country, 'Median wage 2016': median_2016, 'Median wage 2021': median_2021})
wage['Percentage change'] = ((wage['Median wage 2021'] - wage['Median wage 2016']) / wage['Median wage 2016']) * 100
wage['Percentage change'] = wage['Percentage change'].round(2)

wage.sort_values('Percentage change', ascending=False, inplace=True)
wage.reset_index(drop=True, inplace=True)
wage
```

```python
# Plot
plt.figure(figsize=(17, 7))
sns.barplot(x='Percentage change', y='Country', data=wage, palette='viridis')
plt.title('Percentage change in median wage from 2016 to 2021')
plt.xlabel('Percentage change')
plt.ylabel('Country')
for i in range(len(wage)):
    plt.text(wage['Percentage change'][i] + 0.5, i, str(wage['Percentage change'][i]) + "%", ha='left', va='center')
plt.show()
```
